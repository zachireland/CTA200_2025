### Part 1

```python
def f(x):
    """
    Function calculates f(x)=x^3-x^2-1

    Parameters:
    x : float/array-like
        Input value

    Returns:
    y : float/array-like
        Value of f(x)=x^3-x^2-1
    """
    y = x**3 - x**2 - 1
    return y
```

Derivative of $f(x)$ is $\frac{df}{dx}=3x^2-2x$.

```python
def df(x):
    """
    Function calculates derivative of f(x)=x^3-x^2-1

    Parameters:
    x : float/array-like
        Input value

    Returns:
    dydx : float/array-like
           Value of df/dx=3x^2-2x.
    """
    dydx = 3 * x**2 - 2 * x
    return dydx
```

### Part 2

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    Function finds roots (x s.t. f(x)=0) using Newton iteration

    Parameters:
    f : function whose root is being found
    df : derivative of function f
    x0 : initial guess for root (float)
    epsilon : tolerance for convergence (float)
    max_iter : maximum number of iterations allowed (int)

    Returns:
    x : estimated root if method succeeds (float). Returns None if convergence fails
    """
    x = x0  # setting current estimate of root equal to initial guess

    for i in range(max_iter + 1):  # initiating loop for recursion
        if abs(f(x)) < epsilon:
            print(f"Found root in {i} iterations")
            return x

        x = x - f(x) / df(x)  # recursion formula

    print("Iteration failed")
    return None
```

### Part 3

```python
root_1 = newton(f, df, 2)  # calling Newton iteration function on f(x)=x^3-x^2-1 w/ initial guess x0=2 (good guess)
print(root_1)
print(f(root_1))
```

```python
root_2 = newton(f, df, 100)  # initial guess x0=100 (terrible guess)
print(root_2)
print(f(root_2))
```

```python
root_3 = newton(f, df, 2, epsilon=1e-8)  # initial guess x0=2, tolerance epsilon=1e-8
print(root_3)
print(f(root_3))
```

Newton iteration still converges when tolerance is reduced to epsilon=1e-8. For the same initial guess x0=2, it requires only one additional iteration compared with epsilon=1e-6.
